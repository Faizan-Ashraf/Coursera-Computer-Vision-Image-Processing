```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### **Copying Image**

```python
baboon = np.array(Image.open("baboon.png"))
plt.figure(figsize=(3,3))
plt.imshow(baboon)
plt.show()
```

```python
A = baboon.copy()
id(baboon) == id(A)
A[:,:,:] = 0 #[height, width, colorChannel]
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)  #(row,column,position)
plt.imshow(baboon)

plt.subplot(1,2,2) #(row,column,position)
plt.imshow(A)

plt.show()
```

### **Flipping Images**

```python
img = Image.open("cat.png")
plt.figure(figsize = (3,3))
plt.imshow(img)
plt.show()
```

```python
array = np.array(img)
w,h,c = array.shape
print('width, height, C', w, h, c)
```

```python
#Traditional Method
# creating flipped array filled with zeros with same size of original array(w, h, c)
array_flip = np.zeros((w, h, c), dtype=np.uint8)

#assiging row by row
for i,row in enumerate(array):
    array_flip[w - 1 - i, :, :] = row

plt.imshow(array_flip)
plt.show()
```

```python
from PIL import ImageOps
flip_img = ImageOps.flip(img) #flipping
plt.figure(figsize = (7,7))
plt.subplot(1,2,1)
plt.title('Flipped Image')
plt.imshow(flip_img)

mirror_img = ImageOps.mirror(img) #mirroring
plt.subplot(1,2,2)
plt.title('Mirror Image')
plt.imshow(mirror_img)
plt.show()
```

```python
im_flip = img.transpose(1)
plt.imshow(im_flip)
plt.show()
```

```python
flip = {"FLIP_LEFT_RIGHT": Image.FLIP_LEFT_RIGHT,
        "FLIP_TOP_BOTTOM": Image.FLIP_TOP_BOTTOM,
        "ROTATE_90": Image.ROTATE_90,
        "ROTATE_180": Image.ROTATE_180,
        "ROTATE_270": Image.ROTATE_270,
        "TRANSPOSE": Image.TRANSPOSE,
        "TRANSVERSE": Image.TRANSVERSE}

flip["FLIP_LEFT_RIGHT"]
```

```python
for key, values in flip.items():
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("orignal")
    plt.subplot(1,2,2)
    plt.imshow(img.transpose(values))
    plt.title(key)
    plt.show()
```

### **Cropping Images**

```python
upper = 150
lower = 400
crop_top = array[upper: lower,:,:]
plt.imshow(crop_top)
plt.show()
```

```python
left = 150
right = 400
image = Image.open("cat.png")
crop_image = image.crop((left, upper, right, lower))
plt.figure(figsize=(3,3))
plt.imshow(crop_image)
plt.show()
```

### **Changing Specific Pixels**

```python
m_array = np.copy(array)
m_array[upper:lower,left:right:,1:3] = 0
plt.imshow(m_array)
plt.show()
```

```python
#using PIL
from PIL import ImageDraw, Image
img = Image.open("cat.png")

left = 150
right = 400
upper = 150
lower = 400

img_draw = img.copy()
image_fn = ImageDraw.Draw(im = img_draw)

shape = [(left, upper), (right, lower)]
image_fn.rectangle(xy = shape, fill = None, outline = 'red')
plt.imshow(img_draw)
plt.show()
```

### **Writing on Images**

```python
from PIL import ImageFont, ImageDraw

image_fn.text(xy = (0,0), text = "box" ,fill=(0,0,0) )
plt.imshow(img_draw)
plt.show()

```
